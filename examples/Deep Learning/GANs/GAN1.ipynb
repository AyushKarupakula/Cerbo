```
from cerbo.preprocessing import *
from cerbo.DL import *

(X_train, y_train), (X_test, y_test) = load_dataset('fmnist',flatten = False)
X_train = X_train/255.0
```

```
model = GAN(X_train,500)
model.adddenselayer('gen',num_neurons=256)
model.adddenselayer('gen',num_neurons=350)
model.adddenselayer('gen',num_neurons=440)
model.adddenselayer('gen',num_neurons=512)
model.adddenselayer('gen',num_neurons=1024)
model.adddenselayer('gen',activation='tanh',num_neurons=784)


model.adddenselayer('disc',activation='relu',num_neurons=800)
model.adddropout('disc',0.2)

model.adddenselayer('disc',activation='relu',num_neurons=600)
model.adddropout('disc',0.3)


model.adddenselayer('disc',activation='relu',num_neurons=512)
model.adddropout('disc',0.3)

model.adddenselayer('disc',activation='relu',num_neurons=256)
model.adddropout('disc',0.3)

model.adddenselayer('disc',activation='sigmoid',num_neurons=1)

model.train(epochs=300,batch_size=1000,optimizer='adam',lr=.0002)
```
