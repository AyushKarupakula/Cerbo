```
from cerbo.preprocessing import *
from cerbo.ML import Boosting
import numpy as np
(X_train, y_train), (X_test, y_test) = regression_data(X_min=-30, X_max=60, n_samples=2500, n_features=2, noise='medium')

# formatting to use the Boosting() function!
X = np.concatenate((X_train, X_test), axis=0)
Y = np.concatenate((y_train, y_test), axis=0)
data = {
    "X" : X,
    "y" : Y
}

xgb = Boosting("reg", data, split=0.3, algo='xgb', N_estimators=100)
```
