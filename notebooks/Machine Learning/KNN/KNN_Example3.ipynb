```
from cerbo.preprocessing import * # importing everything (*) from the preprocessing submodule
from cerbo.ML import KNN # importing just the KNN() function
import numpy as np # numpy for preprocessing generated data
# we'll be doing classification with the KNN() function
(X_train, y_train), (X_test, y_test) = classification_data(n_clusters=7, n_samples=128)

# formatting to use the KNN() function!
X = np.concatenate((X_train, X_test), axis=0)
Y = np.concatenate((y_train, y_test), axis=0)
data = {
    "X" : X,
    "y" : Y
}

# let's scale our data just for fun 
data = data_scaling(data)

knn, preds = KNN("classification", 7, data, 0.3)
```
