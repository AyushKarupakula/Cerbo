```
!pip install cerbo
from cerbo.preprocessing import *
from cerbo.DL import NeuralNetwork
(X_train, y_train), (X_test, y_test) = load_dataset('mnist',flatten = True)
```

```
#manually entering the layers is more tedious, but gives more customizability


#defining 'model0' as an instance of the 'NeuralNetwork' class
model0 = NeuralNetwork()

#creating the layers for the model
model0.add_input_layer(input_shape=(784,))
model0.add_hidden_layer(neurons = 1024)
model0.add_hidden_layer(neurons = 128, activation_func='tanh')
model0.add_hidden_layer(neurons = 64)
model0.add_output_layer(neurons = 10, activation_func='softmax')

#running the model
model0.run_model(X = X_train, y = y_train, epochs = 5, batch_size = 50)

#evaluating the accuracy of the model on testing data
acc0 = model0.evaluate_accuracy(X_test,y_test)
```

```
#Defining 'model1' as an instance of 'NeuralNetwork' class
model1 = NeuralNetwork()

#Running the model
model1.create_and_run(X = X_train, y = y_train, output_neurons= 10)

#Evaluating model
acc1 = model1.evaluate_accuracy(X_test, y_test)
```
